```python
# Install if needed: pip install pycocotools
from pycocotools.coco import COCO
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
```

### Note for segmentation visualization, export as segmentation mask 

```python
# Load COCO1.1 dataset
coco = COCO(r'C:\Users\sybox\NIJ Detroit\instances_default.json')

# Get basic info
print("="*50)
print("DATASET OVERVIEW")
print("="*50)
print("Categories:", coco.loadCats(coco.getCatIds()))
print("Number of images:", len(coco.getImgIds()))
print("Number of annotations:", len(coco.getAnnIds()))

# Get all image IDs
img_ids = coco.getImgIds()

print("\n" + "="*50)
print("ALL IMAGES INFO")
print("="*50)

# Loop through all images
for i, img_id in enumerate(img_ids):
    img_info = coco.loadImgs(img_id)[0]
    
    # Get annotations for this image
    ann_ids = coco.getAnnIds(imgIds=img_id)
    annotations = coco.loadAnns(ann_ids)
    
    print(f"\nImage {i+1}:")
    print(f"  ID: {img_info['id']}")
    print(f"  Filename: {img_info['file_name']}")
    print(f"  Width: {img_info['width']}")
    print(f"  Height: {img_info['height']}")
    print(f"  Number of annotations: {len(annotations)}")
    
    # Show annotation details
    if annotations:
        print("  Annotations:")
        for j, ann in enumerate(annotations):
            cat_info = coco.loadCats([ann['category_id']])[0]
            bbox = ann['bbox']  # [x, y, width, height]
            area = ann['area']
            print(f"    {j+1}. Category: {cat_info['name']}")
            print(f"       Bbox: x={bbox[0]:.1f}, y={bbox[1]:.1f}, w={bbox[2]:.1f}, h={bbox[3]:.1f}")
            print(f"       Area: {area:.1f}")
            print(f"       ID: {ann['id']}")
    else:
        print("  No annotations")
    print("-" * 30)
```
